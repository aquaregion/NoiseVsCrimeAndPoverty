```python
zips = []
for index, row in df1.iterrows():
    #print(index,row)
    zips.append(row[2].split(','))
```

```python
zips = list(itertools.chain.from_iterable(zips))
```

```python
zipcode = []
for stuff in zips:
    stuff.strip()
    #print stuff.strip()
    zipcode.append(stuff.strip())
```

```python
for stuff in df['Zip Code']:
    #print stuff
    if str(stuff) in zipcode:
        continue
    else:
        df = df[df['Zip Code'] != stuff]
```
